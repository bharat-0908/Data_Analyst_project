```python
import pandas as pd

data = pd.read_csv('d:/python_practice/WA_Fn-UseC_-Telco-Customer-Churn.csv')
df = pd.DataFrame(data)
print(df.head())
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
print(df.columns)
```

```python
def convert(value):
    if value == 1:
        return 'yes'
    else:
        return 'no'
    

df['SeniorCitizen'] = df['SeniorCitizen'].apply(convert)
```

```python
print(df.head())
```

```python
df['customerID'].duplicated().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(3,4))
ax = sns.countplot(x='Churn',data=df)
ax.bar_label(ax.containers[0])
plt.title('Count Of Customers by Churn',fontsize = 13)
plt.show()
```

```python
group =  df.groupby('Churn')['Churn'].count()
print(group)

plt.Figure(figsize=(3,3))
plt.pie(group,labels=group.index,autopct='%.2f%%')
plt.title('Percentage Of Churned Customeres')
plt.show()
```

```python
plt.Figure(figsize=(3,3))
a = sns.countplot(x='gender',data=df,hue='Churn')
a.bar_label(a.containers[0])
plt.legend()
plt.title('Churn by Gender')
plt.show()
```

```python
plt.Figure(figsize=(3,3))
b = sns.countplot(x='SeniorCitizen',data=df,hue='Churn')
for container in b.containers:
    b.bar_label(container)

plt.show()
```
